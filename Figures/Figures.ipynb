```python
import matplotlib.pyplot as plt
import numpy as np

h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wl, T):
    a = 2.0*h*c**2
    b = h*c/(wl*k*T)
    intensity = a/ ( (wl**5) * (np.exp(b) - 1.0) )
    return intensity

# create vector of wavelength values (in meters) for the plots
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

plt.rcParams['figure.figsize'] = [5,3]
for degKelvin in [3000., 4000., 5000., 5777., 6000.]:
    plt.plot(wavelengths*1e9, planck(wavelengths, degKelvin), label='T = '+str(int(degKelvin))+' K') 
plt.grid(True)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Radiance ($kW \cdot sr^{-1} \cdot m^{-2} \cdot nm^{-1}$)')
plt.legend()
plt.title('Blackbody Radiation')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [8, 4] # set the figure size for the plots
# import uniform distribution
from scipy.stats import norm

# plot the probability density function
plt.figure(figsize=(8,4)) # layout to make 1 row, 2 columns of plots, and create to location 2.
x = np.linspace(-5, 5, 100) # generate a list of x-values
y1 = norm.pdf(x, loc=-1, scale=1) # compute the PDF at these x-values
plt.plot(x,y1, label='$\mu=-1,\sigma=1$')
y2 = norm.pdf(x, loc=2, scale=0.75) # compute the PDF at these x-values
plt.plot(x,y2, label='$\mu=2,\sigma=0.75$')
plt.legend()
plt.grid(True)
plt.title('Normal Distribution Probability Density Function', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('$f(x|\mu,\sigma)$', fontsize=15);
```
